 ### 1) Nombre de visites pour chaque fichier différent sur le site Web

```python
%%writefile mapper1.py
#!/usr/bin/env python3
import sys
import re

for line in sys.stdin:
    line = line.strip()

    match = re.search(r'\"(?:GET|POST) ([^ ]+) HTTP', line)

    if match:
        path = match.group(1)
        path = re.sub(r'^http://www.the-associates.co.uk', '', path)

        if path.strip():
            print(f"{path}\t1")
```

```python
!chmod +x mapper1.py
```

```python
%%writefile reducer1.py
#!/usr/bin/env python3
import sys

current_file = None
total_visits = 0

for line in sys.stdin:
    line = line.strip()
    file, count = line.split("\t")

    try:
        count = int(count)
    except ValueError:
        continue

    if current_file and current_file != file:
        print(f"{current_file}\t{total_visits}")
        total_visits = 0

    current_file = file
    total_visits += count

if current_file:
    print(f"{current_file}\t{total_visits}")
```

```python
!chmod +x reducer1.py
```

### 2) Le nombre de visites sur le site effectuées par chaque adresse IP différente.

```python
%%writefile mapper2.py
#!/usr/bin/env python3
import sys
import re

log_pattern = re.compile(r'^(\S+)')

for line in sys.stdin:
    line = line.strip()
    match = log_pattern.match(line)
    if match:
        ip_address = match.group(1)
        print(f"{ip_address}\t1")
```

```python
!chmod +x mapper2.py
```

```python
%%writefile reducer2.py
#!/usr/bin/env python3
import sys

current_ip = None
total_visits = 0

for line in sys.stdin:
    line = line.strip()
    ip_address, count = line.split("\t")

    try:
        count = int(count)
    except ValueError:
        continue

    if current_ip and current_ip != ip_address:
        print(f"{current_ip}\t{total_visits}")
        total_visits = 0

    current_ip = ip_address
    total_visits += count

if current_ip:
    print(f"{current_ip}\t{total_visits}")
```

### 3) Le fichier le plus populaire :

```python
%%writefile mapper3.py
#!/usr/bin/env python3
import sys
import re

for line in sys.stdin:
    line = line.strip()

    # regex pour extraire la partie de la requête (après GET/POST)
    match = re.search(r'\"(GET|POST) (.*?) HTTP', line)
    if match:
        path = match.group(2)

        path = re.sub(r'^http://www.the-associates.co.uk', '', path)

        print(f"{path}\t1")
```

```python
!chmod +x mapper3.py
```

```python
%%writefile reducer3.py
#!/usr/bin/env python3
import sys

current_file = None
total_count = 0

max_file = None
max_count = 0

for line in sys.stdin:
    line = line.strip()

    if not line:
        continue

    if "\t" not in line:
        continue

    try:
        file_path, count = line.split("\t")
        count = int(count)
    except ValueError:
        continue

    if current_file and current_file != file_path:
        if total_count > max_count:
            max_file = current_file
            max_count = total_count

        total_count = 0

    current_file = file_path
    total_count += count

if current_file and total_count > max_count:
    max_file = current_file
    max_count = total_count

print(f"Fichier le plus populaire : {max_file}\nNombre d'occurrences : {max_count}")
```
